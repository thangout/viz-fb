```python
soup = BeautifulSoup(open("../messages.htm"))
soup = BeautifulSoup(soup.encode('UTF-8','ignore'))
```

```python
users = soup.find_all("span", class_="user")
```

```python
threads = soup.find_all("div", class_="thread")
```

```python
count = 0
d = dict()

for e in users:
    key = e.string
    #if(count > 5): 
    #    break
    
    if key in d:
        d[key] += 1
    else:
        d[key] = 1
    
    count += 1
```
